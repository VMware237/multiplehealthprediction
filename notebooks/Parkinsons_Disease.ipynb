```python
# Import libraries and dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
```

```python
parkinsons_df = pd.read_csv('/Users/vaishnavimhaske/Documents/Multiplt_Disease_Prediction/datasets/parkinsons.data')
parkinsons_df.head()
```

```python
# Dropped the name column as it is not needed
parkinsons_df = parkinsons_df.drop(columns=['name'])
```

Based on the above results we can see that the dataset exhibits class imbalance, where the number of positive samples significantly outweighs the negative samples.

```python
# Split the data into features and target variable
X = parkinsons_df.drop(columns=['status'])
Y = parkinsons_df['status']
```

```python
# Handle class imbalance using SMOTE
smote = SMOTE(random_state = 42)
X, Y = smote.fit_resample(X, Y)
```

```python
# Split the data into training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

randomforest = RandomForestClassifier()
randomforest.fit(X_train, Y_train)
```

```python
train_pred = randomforest.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, train_pred)
print("Accuracy score of training data: ", training_data_accuracy)
```

```python
test_pred = randomforest.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_pred)
print("Accuracy score of test data: ", test_data_accuracy)
```

```python
# Model Evaluation
print(classification_report(Y_test, test_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
```

```python
grid_search = GridSearchCV(estimator=randomforest, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, Y_train)
```

```python
# Best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
print("Best parameters found: ", best_params)
print("Best cross-validation score: ", best_score)
```

```python
# Train and evaluate the model with best parameters
best_randomforest = grid_search.best_estimator_
y_pred_best_randomforest = best_randomforest.predict(X_test)
best_randomforest_accuracy = accuracy_score(Y_test, y_pred_best_randomforest)
print("Best Random Forest Accuracy: ", best_randomforest_accuracy)
```

```python
# Saving the model and the scaler
import pickle
filename = 'Parkinsons_Prediction.sav'
scaler_filename = 'scaler_parkinsons.sav'
pickle.dump(best_randomforest, open(filename, 'wb'))
pickle.dump(scaler, open(scaler_filename, 'wb'))
```

```python
# Loading the model and the scaler
loaded_model = pickle.load(open(filename, 'rb'))
loaded_scaler = pickle.load(open(scaler_filename, 'rb'))
```
